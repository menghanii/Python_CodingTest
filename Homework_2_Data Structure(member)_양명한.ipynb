### [HW2] 8. 성별  회당 평균 구매 금액을 비교   
- 고객별 회당 평균 구매금액을 구함
- 성별로  회당 평균 구매금액의 평균을 구함

```python
# 고객별 회당 평균 구매금액

# name, count, sales

name = []
count = []
sales = []
average = []
dic = dict()

for i in mem_card:
    name.append(i["name"]) # key가 name인 것들만 뽑아서 순서대로 name 리스트에 저장
    count.append(i["count"])
    sales.append(i["sales"])

for q in range(30):
    avg = sales[q] / count[q]
    average.append(avg)
    dic[name[q]] = average[q]
    
dic
```

```python
# 성별 회당 평균 구매금액

male_list = []
female_list = []

male_count = []
female_count = []

male_sale_list = []
female_sale_list = []

for i in mem_list:
    if i[1] == 'F':
        female_list.append(i)
    else:
        male_list.append(i)
    
for m in male_list:
    male_count.append(m[-1])
    male_sale_list.append(m[-2])
    
for f in female_list:
    female_count.append(f[-1])
    female_sale_list.append(f[-2])

a = sum(male_sale_list)
b = sum(male_count)
c = sum(female_sale_list)
d = sum(female_count)

print("남성 회당 평균 구매금액 >>", a / b)
print("여성 회당 평균 구매금액 >>", c / d)
```
