```python
# 11021번 문제

T = int(input())
for i in range(1, T + 1):
    A, B = map(int, input().split())
    print("Case #"+ str(i) + ":", (A+B))
```

```python
# 10871번 문제

A, X = map(int, input().split())
less_X = input().split()

for i in range(A):
    if int(less_X[i]) < X:
        print(int(less_X[i]), end=' ')
```

```python
# 1110번 문제

N = int(input())
N_test = N + 0
num_list = []
counter = 0

while True:
    if N < 10:
        num_list = [0, N]
    else:
        str_N = str(N)
        num_list = [int(str(N)[0]), int(str(N)[1])]

    new_num = 10 * num_list[-1] + int(str(num_list[0]+num_list[1])[-1])
    N = new_num
    counter += 1
    
    if N == N_test:
        break
        
print(counter)
```
